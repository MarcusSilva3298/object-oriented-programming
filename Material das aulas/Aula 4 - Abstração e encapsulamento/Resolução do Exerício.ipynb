# Máquina de Café 

Primeiro identificamos os atributos da classe
 * **total** : Total de dinheiro armazenado na máquina
 * **dinheiro** : Dinheiro atualmente inserido pelo usuário (para comprar o café)
 * **troco** : Troco 
 * **cafe** : O café está pronto ?
 * **valorCafe**: Valor de um café

Depois identificamos os métodos da classe:

 * **moeda5** : Inserir uma moeda de 5
 * **moeda10** : Inserir uma moeda de 10
 * **pegarCafe/Troco** : pegar o café/troco (quando estiver pronto)
 * **consultar** : consultar o estado da máquina

Para implementar esses métodos, podemos alguns métodos **privados**:
 * **moeda** : Método privado para inserir qualquer moeda/valor

```python
class Maquina:
    '''Controlador de uma máquina de café'''
    
    def __init__(self, valorCafe=50):
        #Atributos Públicos
        self.valorCafe = valorCafe  

        #Atributos Privados
        self.__total = 0 #Total de dinheiro
        self.__dinheiro = 0 #Total atual (do usuário atual)
        self.__troco = 0 #Troco
        self.__cafe = False #Determina se o café está pronto
        
    def __str__(self):
        #Mais um exemplo do operador ternario
        scafe = 'Café Pronto!' if self.__cafe else 'Sem Café'
        return f'''
                --------------------------------
                Estado da máquina
                --------------------------------
                *  Total: {self.__total}.
                *  Dinheiro usuário: {self.__dinheiro} / {self.valorCafe}.
                *  Café: {scafe}.
                *  Troco: {self.__troco}. 
                --------------------------------
                '''
        
    #Método privado    
    def __moeda(self, valor):
        '''Inserir uma moeda'''
        if self.__dinheiro >= self.valorCafe:
            print('Primeiro deve pegar o Café')
        else:
            # Incrementar o dinheiro na máquina
            self.__dinheiro += valor
            self.__total += valor
            
            #Determinar se o usuário já pagou suficiente
            if self.__dinheiro >= self.valorCafe:
                self.__cafe = True
                self.__troco = self.__dinheiro - self.valorCafe
                
    def moeda5(self):
        '''Inserir moeda de 5'''
        self.__moeda(5)
        
    def moeda10(self):
        '''Inserir moeda de 10'''
        self.__moeda(10)
        
    #Implementamos métodos GET para o dinheiro, troco e café
    @property
    def total(self):
        return self.__total
    
    @property
    def dinheiro(self):
        return self.__dinheiro
    
    @property
    def troco(self):
        return self.__troco
    
    @property
    def cafe(self):
        return self.__cafe
    # FIM Métodos GET
    
    def pegarCafe(self):
        '''Retorna o troco'''
        if not self.__cafe:
            print('Café não disponível')
            return 0
        else:
            # Note que a máquina volta para o estado inicial (sem café, sem troco e sem dinheiro)
            self.__cafe = False
            valor = self.__troco
            self.__total -= valor
            self.__troco = 0
            self.__dinheiro = 0
            return valor
        
    
#TEST
M = Maquina()
print(M)
M.moeda5()
print(M)
M.moeda10()
M.moeda10()
M.moeda10()
M.moeda10()
print(M)
M.moeda10()
print(M)
#Não aceita mais moedas
M.moeda5()
#Pegamos o café
troco = M.pegarCafe()
print(f'Troco: {troco}')
print(M)
#Utilizar o método get para o total de vendas
print(f"Total vendas: {M.total}")
print("Comprar 10 cafés")
for i in range(0,10):
    for j in range(0,5):
        M.moeda10()
    print(M)
    M.pegarCafe()

print(f"Total vandas: {M.total}")    
```
